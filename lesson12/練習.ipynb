```python
import random
import matplotlib.pyplot as plt

#figure  
figure = plt.figure(figsize=(10,4)) #figsize(x,y)
#figureaxes
axes = figure.add_subplot(1, 1, 1)

days = [num for num in range(1,31)]
celsius = [round(random.uniform(28.5, 35.2),1) for _ in range(1,31)]

axes.grid(axis='both')  #
axes.set_title('july weather') #
axes.set_xlabel('Days') #x
axes.set_ylabel('Celsius') #y
axes.set_xticks(days) #x 1~30
axes.plot(days, celsius,'ro--')
#
plt.show()
```
